```python
from spp import *
from pandas import read_csv
```

```python
datas = read_csv('sharedata.csv', usecols=[0], engine='python')
dataset = datas.values
dataset = dataset.astype('float32')

dataset=normalize(dataset)
train_test_ratio=0.7
train_size = int(len(dataset) * train_test_ratio)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]
look_back = 1
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)
trainX,testX=shape_data(trainX,testX)
model=build_model(1,50,100,0.2)
train_model(model,trainX,trainY)
```

```python
no_of_predictions=10
points=predict(model,testX,no_of_predictions)
plot(dataset,points,no_of_predictions)
```
